```python
import seaborn as sns
%matplotlib inline
sns.set(font_scale=1.5)
```

```python
data = sns.load_dataset("flights").pivot(index="month", columns="year", values="passengers")
data.head()
```

```python
# Set the size of figure
sns.set_context({"figure.figsize": (8, 8)})
# demo
sns.heatmap(data=data, square=True)
```

```python
sns.heatmap(data=data,cmap="RdBu_r", vmax=600, vmin=100, square=True)
```

```python
# fmt="d" ---> Integer 
sns.heatmap(data=data,cmap="RdBu_r", vmax=600, vmin=100, square=True, annot=True, fmt="d")
```
